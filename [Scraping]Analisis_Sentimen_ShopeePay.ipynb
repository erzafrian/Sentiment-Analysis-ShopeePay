```python
from google_play_scraper import app, reviews, Sort # pustaka untuk mengakses ulasan dan informasi aplikasi dari Google Play Store

import pandas as pd # manipulasi dan analisis data
pd.options.mode.chained_assignment = None # menonaktifkan peringatan chaining
```

**Scraping Dataset**

```python
# Mengambil 10.000 ulasan dari aplikasi Ajaib dengan ID 'ajaib.co.id' di Google Play Store
scrapreview = reviews(
    'com.shopeepay.id',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=25000,
)
```

```python
# membuat dataframe dari hasil scrapreview
scrap_df = pd.DataFrame(scrapreview[0])

jumlah_ulasan, jumlah_kolom = scrap_df.shape
print(f"Jumlah ulasan: {jumlah_ulasan}")
print(f"Jumlah kolom: {jumlah_kolom}")
```

```python
# menampilkan informasi dataset scraping
scrap_df.info()
```

```python
# menampilkan 5 baris pertama dari dataset scraping
scrap_df.head()
```

```python
# menyimpan DataFrame ke file CSV
scrap_df.to_csv('shopeepay_reviews4.csv', index=False)

print(f"\nSuccessfully scraped {jumlah_ulasan} reviews and saved to 'shopeepay_reviews4.csv'")
```
